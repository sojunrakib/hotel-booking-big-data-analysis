```python
# Plot histogram to see distribution of lead_time
plt.figure(figsize=(8,5))

plt.hist(data['lead_time'], bins=30)

plt.title("Lead Time Distribution")
plt.xlabel("Lead Time")
plt.ylabel("Frequency")

plt.show()
```

```python
# Show number of bookings for each hotel type
data['hotel'].value_counts().plot(kind='bar')

plt.title("Hotel Type Distribution")
plt.xlabel("Hotel Type")
plt.ylabel("Number of Bookings")

plt.show()
```

```python
# Visualize how many bookings were cancelled vs not cancelled
sns.countplot(x='is_canceled', data=data)

plt.title("Booking Cancellation Distribution")

plt.show()
```

# Big Data Analysis Report

## Hotel Booking Demand Dataset

---

# 1. Dataset Description

**Dataset Source:** Kaggle
Dataset Link:
[https://www.kaggle.com/datasets/jessemostipak/hotel-booking-demand](https://www.kaggle.com/datasets/jessemostipak/hotel-booking-demand)

This dataset contains information about hotel bookings. It includes details about customers, booking time, number of guests, and booking cancellation.

**Dataset Information**

* **Number of Rows:** 119,390
* **Number of Columns:** 32
* **Data Types:**

  * Numerical data (numbers)
  * Categorical data (text)

Examples:

Numerical columns:
lead_time, adults, children, adr

Categorical columns:
hotel, country, market_segment

---

# 2. Business Problem

Hotels receive many booking requests every year. But many bookings get cancelled before the arrival date. This can cause problems for hotel management and reduce revenue.

The goal of this analysis is to understand:

* booking patterns
* customer behavior
* cancellation trends

This information can help hotels improve their booking system and reduce cancellations.

---

# 3. Handling Missing Values

Some columns in the dataset had missing values. For example:

* country
* children
* agent
* company

To fix this problem, missing values were replaced.

* Numerical missing values were replaced with **0**
* Categorical missing values were replaced with **"Unknown"**

After this step, the dataset had **no missing values**.

---

# 5. Outliers, Patterns and Insights

During the analysis some interesting observations were found.

**Outliers**

The ADR (Average Daily Rate) column shows some extreme values which are considered outliers.

**Patterns**

* City hotels have more bookings than resort hotels.
* Higher lead time often leads to higher cancellation.

**Trends**

* Some market segments bring more customers.
* Number of guests (adults and children) affects booking patterns.

These findings help understand hotel booking behavior.

Training data is used to learn patterns from the dataset.
Testing data is used to check if the model works well on new data.

This helps make the analysis more reliable.
